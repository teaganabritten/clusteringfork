**Q1.** This is a question about clustering. We want to investigate how adjusting the "noisiness" of the data impacts the quality of the algorithm and the difficulty of picking $k$.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

```
import numpy as np
```

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
# 2

import seaborn as sns
import matplotlib.pyplot as plt

dfs = [df0_125, df0_25, df0_5, df1_0, df2_0]

def plotdfs(dfs):
    for df in dfs:
        sns.scatterplot(df, x = 'x1', y = 'x2', hue = 'group')
        plt.show()

plotdfs(dfs)
````

The clusters become less and less clear as the noise increases, eventually reaching a point where the clusters are entirely indistinguishable.

````python
# 3

from sklearn.cluster import KMeans

k_bar = 10
scree_data = []  # To store SSE for all datasets
noise = [0.125, 0.25, 0.5, 1.0, 2.0]

for i, (df, noise) in enumerate(zip(dfs, noise)):
    SSE = []
    for k in range(1, k_bar + 1):
        model = KMeans(n_clusters=k, max_iter=300, n_init=10, random_state=0)
        model.fit(df[['x1', 'x2']]) 
        SSE.append(model.inertia_)
    scree_data.append((noise, SSE))

plt.figure(figsize=(10, 6))
for noise, SSE in scree_data:
    plt.plot(range(1, k_bar + 1), SSE, label=f'Noise={noise}')
plt.title('Scree Plots for Different Noise Levels')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.legend()
plt.show()
````

The most clear observation I can make is that the SSE increases as the noise increases. The graphs also clear to have a less and less clear elbow point as the noise increases, making it more difficult to pick a K to cluster the already noisy data. 

4. The intuition of the elbow is to find the point at which the SSE decreases much less by adding another cluster. The idea is that the 'elbow' represents a change in the derivative strong enough that the value K at which that occurs returns the optimal benefit of clustering without overfitting or underfitting the model.

````python
# 1

columns = ['Raised', 'Spent', 'Cash on Hand', 'Debts']
for col in columns:
    ec[col] = ec[col].str.replace('$', '', regex=False).str.replace(',', '', regex=False)
    ec[col] = pd.to_numeric(ec[col], errors='coerce')
'''
I removed the dollar signs and commas from each column to leave just a number.
I then made each column numeric to allow for their use in clustering.
'''
````

````python
# 2

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ec[['Raised', 'Spent']] = scaler.fit_transform(ec[['Raised', 'Spent']])

k_bar2 = 10
SSE2 = []

for k in range(1, k_bar2 + 1):
    model = KMeans(n_clusters=k, max_iter=300, n_init=10, random_state=0)
    model.fit(ec[['Raised', 'Spent']])
    SSE2.append(model.inertia_)


plt.plot(range(1, k_bar2 + 1), SSE2, marker='o')
````

````python
optimal_k = 3  
model = KMeans(n_clusters=optimal_k, max_iter=300, n_init=10, random_state=0)
ec['Cluster'] = model.fit_predict(ec[['Raised', 'Spent']])

sns.scatterplot(data=ec, x='Raised', y='Spent', hue='Cluster', palette='bright')
````
